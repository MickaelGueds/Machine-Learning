    13 VOcê conseguiria a variação percentual ano a ano do rj 

```python

DFGasolina['MES'] = DFGasolina["DATA FINAL"].apply(lambda x: x.month)

DFRio = DFGasolina[DFGasolina['ESTADO'] == 'RIO DE JANEIRO' ]
```

```python
DFMES_RIO = DFRio.groupby('ANO-MES')[['PREÇO MÉDIO REVENDA',"MES"]].last()

#Aqui vai ser complicado para criar a variação, admito que não consegui e tive que ver a aul, mas tem duas funções extras que ajudam,
#O shift() que serve para descolocar os valores na unidade e então voce faz o calculo para porcentagem, VEJA:

(DFMES_RIO[DFMES_RIO["MES"]==12]/DFMES_RIO[DFMES_RIO["MES"]==12].shift(1) - 1) *100 #calculo da variação percentual 
```
